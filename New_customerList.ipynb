```python
df['last_name'].isna().sum()
```

```python
df.loc[df['DOB'].isna()]
```

```python
#as the mean of the missing values of DOB is less than 5%. we can remove it.
index_pop = df[df['DOB'].isna()].index #get the indexes of the DOB rows where its null
index_pop
```

```python
df.drop(index = index_pop, axis = 0, inplace = True)
```

## Drop the unnamed columns


```python
df['DOB'].isnull().sum()
```

```python
df.drop(columns=['Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20'], inplace = True)
```

## Create a age column as a new feature

```python
today_date =date.today()
today_date
```

```python
# Function to find the age of customers as of today.

def age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

df['Age'] = df['DOB'].apply(age)
```

```python
df.head()
```

```python
#Descriptive statics of the age columns
df['Age'].describe()
```

```python
plt.figure(figsize=(15,9))
sns.distplot(df['Age'], kde=False, bins = 50)
```

```python
#creating a age group
import math
df['Age_Group'] = df['Age'].apply(lambda x: (math.floor(x/10)+1)*10)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(df['Age_Group'], kde = False, bins = 50)
```
